```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common import exceptions
import time
import logging
import datetime

driver = webdriver.PhantomJS('./phantomjs')

# 코드 인코딩 변경

# 기후변화 단어들
s_list = ['강추위','결빙','무더위','열대야','열섬','열파','온난','온실가스','이상고온','이상기온','이상저온','폭염','한파',
          '혹서','혹한','짱춥','짱덥','졸춥','졸덥','넙덥','넘춥','너무춥','너무덥','개춥','개덥','대설','산성비','우박','장마',
          '적설','집중강우','집중호우','폭설','폭우','홍수','황사비','눈난리','눈쓰레기','눈폭탄','물난리','비폭탄','흙비','흙탕물비'
          '가뭄','사막화','산불','산사태','열대림파괴','지진','토지황폐화','화산폭발','갈라진땅','메마른땅','산폭팔','찢어진땅',
          '흔들리는땅','녹조','라니냐','쓰나미','엘니뇨','적조','침수','파랑','풍랑','풍수해','해랑','해수면','해일','괴물파도','큰파도'
]


dt = str(datetime.datetime.today().date())
src_list = []
text_list = []
key_list = []
date_list = []
for s in s_list:
    print(s)
    key = str(s.encode('utf-8'))
    key = key.replace(r"b'",' ')
    key = key.replace(r"'",' ')
    key = key.replace(r"\x","%")
    key = key.strip()
    print(key)
    driver.implicitly_wait(3)

    driver.get('https://search.naver.com/search.naver?where=realtime&sm=tab_rea&query=' + key)
    #print(driver)

    try:
        # 더보기가 있을경우 최대한 활성화
        for c in range(100):
            driver.find_element_by_css_selector('#realTimeSearchBody > div.rt_wrap.rt_typ2 > div.bt_more > a').click()
            time.sleep(2)

    except Exception as e:
        content_list = driver.find_elements_by_css_selector('#realTimeSearchBody > div.rt_wrap.rt_typ2 > div:nth-of-type(1) > ul:nth-of-type(1) > li > dl > dd:nth-of-type(2)')       
        for i in content_list:
            key_list.append(s)
            date_list.append(dt)
            try:
                i.find_element(By.TAG_NAME, 'span')  # 트위터
                src_list.append('트위터')
                text_list.append(i.text)
            except exceptions.NoSuchElementException:
                src_list.append('네이버')
                text_list.append(i.text)
    
    print('content_list size :',len(content_list))
    
driver.quit()
```

```python
result_mat = [date_list,src_list,key_list, text_list]
```

```python
import pandas as pd
result_df = pd.DataFrame(result_mat)
result_df = result_df.T
```

```python
filename = str(datetime.datetime.today().date()) + "_"+ str(datetime.datetime.today().hour) + ".txt"
```

```python
result_df.to_csv("/media/Data/Social_Comments/" + filename,index=None,columns=None,header=False)
```
